## 1. a. What is PCA? What does a PCA do?

##    b. List down the steps of a PCA algorithm.

**a)**. PCA stands for Principal Component Analysis. It is a statistical technique used for data analysis and dimensionality reduction. The goal of PCA is to identify the underlying structure in high-dimensional data and reduce the dimensionality while retaining as much information as possible. PCA achieves this by finding a new set of orthogonal variables called principal components, which are linear combinations of the original variables that capture the maximum amount of variation in the data.

**b**. The steps of a PCA algorithm are as follows:

Standardize the data: PCA requires that the data is standardized to have a mean of zero and a variance of one.

Calculate the covariance matrix: Calculate the covariance matrix of the standardized data. The covariance matrix represents the relationships between the variables in the data.

Calculate the eigenvectors and eigenvalues: Find the eigenvectors and eigenvalues of the covariance matrix. The eigenvectors represent the principal components, while the eigenvalues represent the amount of variance explained by each principal component.

Sort the eigenvectors by eigenvalue: Sort the eigenvectors in descending order based on their eigenvalues. The eigenvectors with the highest eigenvalues explain the most variance in the data and are the most important principal components.

Choose the number of principal components: Decide on the number of principal components to retain based on the percentage of variance explained or by setting a threshold for the eigenvalues.

Project the data onto the new principal components: Project the data onto the new principal components to obtain the transformed data with reduced dimensionality.

Interpret the results: Interpret the results by analyzing the principal components and the transformed data. The principal components can be used to identify the underlying structure in the data, and the transformed data can be used for further analysis or modeling.

## 3. What are the differences in Anomalies for Uniform Distribution and Normal Distribution in One-Dimensional Data?

he main difference between anomalies or outliers for uniform distribution and normal distribution in one-dimensional data lies in their shape and location.

In a uniform distribution, the data is evenly spread out between a minimum and maximum value. Anomalies in a uniform distribution may appear as values that fall outside the minimum and maximum range or values that are significantly different from the surrounding values. In general, anomalies in a uniform distribution are rare and are typically easier to detect because they are located in the tails of the distribution.

On the other hand, in a normal distribution, the data is typically bell-shaped, with the majority of the data clustered around the mean value. Anomalies in a normal distribution may appear as values that are significantly higher or lower than the mean value or as values that fall outside a certain range, typically defined by the standard deviation of the data. In general, anomalies in a normal distribution are more common than in a uniform distribution and can be more difficult to detect because they are located in the center of the distribution.

In summary, the differences in anomalies or outliers for uniform distribution and normal distribution in one-dimensional data lie in their shape, location, and frequency. Anomalies in a uniform distribution are rare and located in the tails of the distribution, while anomalies in a normal distribution are more common and located in the center of the distribution.

## 4. a. What are the Swamping and Masking problems in Anomaly Detection? 

## b. What is the Change Detection problem in Anomaly Detection? 

**a.)**Swamping and masking are two common problems in anomaly detection:

Swamping problem: Swamping occurs when a large number of normal data points are incorrectly classified as anomalies, leading to a high false positive rate. This can happen when the anomaly detection algorithm is not tuned properly, or when the data contains a high degree of noise or outliers that make it difficult to distinguish between normal and anomalous data points.

Masking problem: Masking occurs when an anomalous data point is masked or hidden by the surrounding normal data points, leading to a high false negative rate. This can happen when the anomaly detection algorithm is not sensitive enough to detect subtle anomalies or when the data contains a large number of similar normal data points that make it difficult to identify the anomaly.

**b.** The change detection problem in anomaly detection refers to the detection of changes in the statistical properties of a system or process that may indicate the presence of anomalies. Change detection algorithms monitor the data stream over time and look for significant deviations or changes from the expected or baseline behavior. These changes may be caused by anomalies such as faults, errors, or cyber-attacks, and detecting them early can help prevent potential damage or loss.

Change detection algorithms can be used in various applications, such as intrusion detection, fraud detection, and condition monitoring, where the goal is to identify anomalies or changes in the data that may indicate a problem or threat. However, change detection algorithms can also suffer from false positives or false negatives if the threshold for detecting changes is not set properly or if the data contains a high degree of noise or variability. Therefore, it is important to validate and fine-tune the change detection algorithm to minimize the rate of false alarms and missed detections.





## 5. What are the steps to follow when building a text classification system?

Building a text classification system involves several steps, and the specific approach may vary depending on the problem and the available data. However, the general steps to follow when building a text classification system are as follows:

Define the problem and the objective: The first step is to define the problem and the objective of the text classification system. This involves understanding the business or research problem and identifying the categories or labels that the text data should be classified into.

Gather and preprocess the data: The second step is to gather and preprocess the text data. This involves collecting a representative sample of the data, cleaning and normalizing the text, and converting it into a format that can be processed by the machine learning algorithms.

Feature extraction: The third step is to extract relevant features from the text data. This involves selecting appropriate text representation techniques such as bag-of-words, tf-idf, or word embeddings, and transforming the text data into numerical features that can be used as input to the machine learning algorithms.

Select and train a machine learning algorithm: The fourth step is to select an appropriate machine learning algorithm for the text classification task, such as logistic regression, support vector machines, or neural networks.
